```python
#Exercise 2
import sympy as sp

x0 = 0
y0 = 1
h = 0.1

x, y = sp.symbols('x y')
f = y**2 - x**3

# 1. Taylor Series Method of Order 2
f1 = f

df_dx = sp.diff(f1, x)
df_dy = sp.diff(f1, y)

f2 = df_dx + df_dy * f1

f1_val = f1.subs({x: x0, y: y0})
f2_val = f2.subs({x: x0, y: y0})

y_taylor = y0 + h * f1_val + (h**2 / 2) * f2_val

print("1. Taylor Series Method (Order 2):")
print(f"y(0.1) ≈ {float(y_taylor):.6f}")

#  2. Runge–Kutta Method of Order 4
def f_numeric(x, y):
    return y**2 - x**3

k1 = f_numeric(x0, y0)
k2 = f_numeric(x0 + h/2, y0 + h/2 * k1)
k3 = f_numeric(x0 + h/2, y0 + h/2 * k2)
k4 = f_numeric(x0 + h, y0 + h * k3)

y_rk4 = y0 + (h/6)*(k1 + 2*k2 + 2*k3 + k4)

print("\n2. Runge–Kutta Method (Order 4):")
print(f"y(0.1) ≈ {y_rk4:.6f}")
```
